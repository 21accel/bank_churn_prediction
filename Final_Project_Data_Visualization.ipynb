```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbrn
```

##Main Data

```python
data = pd.read_csv("https://raw.githubusercontent.com/21accel/bank_churn_prediction/main/BankChurners.csv")
```

```python
data.drop(columns=['CLIENTNUM'], axis=1, inplace=True)
data
```

##Attrition Flag

```python
data.value_counts('Attrition_Flag')
```

```python
# get count existing user and attrited user
existing = data.query("Attrition_Flag == 'Existing Customer'")
existing = existing['Gender'].count()
attrited = data.query("Attrition_Flag == 'Attrited Customer'")
attrited = attrited['Gender'].count()
attrition_Flag = pd.DataFrame(data=[existing,attrited], index=['Existing Customer','Attrited Customer'],columns=['Counts'])
attrition_Flag_T = attrition_Flag.T
attrition_Flag_T
```

```python
label = ['Existing Customer','Attrited Customer']
counts = [existing, attrited]
# bar chart for Attrition_Flag
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

plt.figure(figsize = (10,5))
# making the bar chart on the data
plt.bar(label, counts)
  
# calling the function to add value labels
addlabels(label, counts)

# giving title to the plot
plt.title("Attrited Flag")
plt.show()
```

```python
pie, ax = plt.subplots(figsize=[10,10])
plt.pie(x=counts, autopct="%.1f%%", pctdistance=1.1, labeldistance=1.25)
plt.legend(labels=label)
plt.title("Attrited Flag", fontsize=20);
pie.show()
```
